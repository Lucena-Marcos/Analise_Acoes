**Customizando faixas de Eixos (range)**

```
# importa novamente o plotly graph
import plotly.graph_objs as go

# usamos novamente o Scatter
close = go.Scatter(
    # definimos o eixo x como data
    x = COGN3.Date,
    # o eixo y como Fechamento
    y = COGN3.Close,
    name = 'COGN3 Close',
    # definimos a cor da linha em HTML
    line = dict(color = '#17becf'),
    # Definimos a opacidade da linha 
    opacity = 0.8)

# atribuimos tudo isso a variavel data
data = [close]

# agora iniciamos o layout do grafico
layout = dict (
    # Definimos um titulo para o grafico
    title = 'Faixa de Data Customizada',
    # deficnimos onde o titulo ficará, no caso 0.5 é central
    title_x = 0.5,
    xaxis = dict (
        range = ['2017-01-01','2017-12-31'])
)
# Definimos uma figura
fig = dict (data=data, layout=layout)
# chamamos novamente o iplot
py.iplot(fig, filename = 'Analise 2017')
```

**Rangeslider and Rangeslector**

```
# novamente importando o plotly.graph porem não a necessidade novamente
import plotly.graph_objs as go

# copiarei o programa acima e só adicionarei outras partes
close = go.Scatter(
     x = COGN3.Date,
     y = COGN3.Close,
     name = 'COGN3 Close',
     line = dict (color = '#17BECF'),
     opacity = 0.8)
 
data = [close]

layout = dict (
     title = 'Série com RangeSlide e botões',
     title_x = 0.5,
     xaxis = dict (
         # Dicionário rangeselector
         rangeselector = dict (
             # lista de Botões
             buttons = list ([
                dict (count = 1,
                      # denine a quantidade de mês
                      label = '1m',
                      step = 'month',
                      # backward quer dizer que é do mes escolhido para trás
                      stepmode = 'backward'),
                dict (count = 6,
                      label = '6m',
                      step = 'month',
                      stepmode = 'backward'),
                dict (count = 2,
                      label = '1y',
                      step = 'year',
                      stepmode = 'backward'),
                dict (step = 'all')
             ])
         ),
         # slider possibilita fazer a seleção de valores no slide 
         rangeslider = dict (
             visible = True
         ),
         type = 'date'
    )
)
''' The 'step' property is an enumeration that may be specified as:
      - One of the following enumeration values:
      ['month', 'year', 'day', 'hour', 'minute', 'second','all']'''
```

# **TRABALHANDO COM CANDLESTICK**


```
# mostra a variação dos preçõs
trace = go.Candlestick(x = COGN3['Date'],
                       open = COGN3['Open'],
                       high = COGN3['High'],
                       low = COGN3['Low'],
                       close = COGN3['Close'])


data = [trace]
py.iplot(data, filename = 'Simple_Candlestick')
```

**Sem Rangeslisder**

**Custimizando os Candlestick**

```
# candlestick deixa a variação da acção em formato de vela
# se a ação ficar positiva num dia ela fica verde
# se for negativa, fica vermelha

trace = go.Candlestick(x = COGN3['Date'],
                     open = COGN3['Open'],# o valor dentro dos colchetes é o que está no arquivo
                     high = COGN3['High'],
                     low = COGN3['Low'],
                     close = COGN3['Close'],
                     #definimos as cores que a subida e descida de ações deve ter
                     increasing = dict (line = dict (color = '#17BECF')),
                     decreasing = dict (line = dict (color = '#7f7f7f'))
                     )
# Começamdo o Layout
layout = go.Layout(
    #Definindo o Tema do Grafico
    title = 'Preços em Candlesticks',
    # Definindo o titulo no centro
    title_x = 0.5,

    xaxis = dict (
        rangeslider = dict (
            visible = False
        )
    )
)

data = [trace]

# 
fig = go.Figure(data = data, layout = layout)
# Comando para Mostrar o Grafico
py.iplot(fig)

```

# **TRABALHANDO COM MÉDIAS**

```
COGN3.head()
```

```
# MÉDIA SIMPLES DE 9 DIAS
COGN3['MM_3'] = COGN3.Close.rolling(window = 3).mean()

# MÉDIA SIMPLES DE 17 DIAS
COGN3['MM_7'] = COGN3.Close.rolling(window = 7).mean()

# MÉDIA SIMPLES DE 1 ANO
COGN3['MM_9'] = COGN3.Close.rolling(window = 9).mean()
```

```
COGN3.head(20)
```

**PLOTA AS MÉDIAS E OS PREÇOS**

```
# REFENCIANDO OS DADOS DE FECHAMENTO
close = go.Scatter(
    x = COGN3.Date,
    y = COGN3.Close,
    name = 'COGN3 Fechamento',
    line = dict (color = '#330000'),
    opacity = 0.8)

MM_3 = go.Scatter(
    x = COGN3.Date,
    y = COGN3['MM_3'],
    name = 'Média Movel 9 pregões',
    line = dict (color = '#B2FF66'),
    opacity = 0.8)

MM_7 = go.Scatter(
    x = COGN3.Date,
    y = COGN3['MM_7'],
    name = 'Média Movel 7 Pregões',
    line = dict (color = '#FF00FF'),
    opacity = 0.8)
  
MM_9 = go.Scatter(
    x = COGN3.Date,
    y = COGN3['MM_9'],
    name = 'MÉDIA MOVEL EM 1 ANO',
    line = dict (color = '#000000'),
    opacity = 0.8)


# PARA EXECULTAR TODOS, COLOQUE EM UMA VARIAVEL
data = [close,MM_3,MM_7,MM_9]
  
py.iplot(data)
```
